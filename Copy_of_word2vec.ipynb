```python
import pandas as pd
import numpy as np
import os
import gensim
import nltk
```

```python
nltk.download("all")
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
import os
```

```python
story=[]
```

```python
for file in os.listdir("/content/data"):
  print(file)
```

```python
f=open("/content/data/001ssb[1].txt")
corpus=f.read()
```

```python
sent_cor=sent_tokenize(corpus)
```

```python
for i in sent_cor:
  print(simple_preprocess(i))
```

```python
for file in os.listdir("/content/data"):
  file_path=os.path.join("/content/data",file)
  with open(file_path,encoding='unicode_escape') as f:
    corpus=f.read()
  raw_sent=sent_tokenize(corpus)
  for i in raw_sent:
    story.append(simple_preprocess(i))
```

```python
story
```

```python
len(story)
```

```python
story[0]
```

```python
model=gensim.models.Word2Vec(window=10,min_count=5,vector_size=150)
```

```python
model.build_vocab(story)
```

```python
model.corpus_count
```

```python
model.epochs
```

```python
model.corpus_total_words
```

```python
model.min_count
```

```python
model.train(story,total_examples=model.corpus_count,epochs=5)
```

```python
len(model.wv['king'])
```

```python
model.wv.most_similar("king")
```

```python
model.wv.get_normed_vectors().shape
```

```python
len(model.wv.index_to_key)
```

```python
model.save("word2vec.model")
```
